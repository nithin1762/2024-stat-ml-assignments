```python
import pandas as pd
data = pd.read_csv('/content/breast_cancer_survival.csv')
data
```

```python
data=data.dropna()
data = data.replace({'FEMALE':0, 'MALE':1, 'Negative':0, 'Positive':1, 'Alive':1, 'Dead':0, 'I':1, 'II':2,'III':3, 'Infiltrating Ductal Carcinoma':1, 'Infiltrating Lobular Carcinoma':2, 'Mucinous Carcinoma':3, 'Lumpectomy':1, 'Modified Radical Mastectomy':2, 'Simple Mastectomy':3, 'Other':4})
data
```

```python
x = data.drop(['Patient_Status', 'Date_of_Surgery', 'Date_of_Last_Visit'], axis=1) #features
y = data['Patient_Status'] #target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svm = SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
```

```python
lr_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", lr_accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
```

```python
svm_accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", svm_accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
knn_accuracy = knn.score(x_test, y_test)
```

```python
print(knn_accuracy)
```

```python
import numpy as np
from sklearn.utils import resample
```

```python
def bootstrap_model(model, x, y, n_iterations=1000):
  accuracy_scores = []
  for _ in range(n_iterations):
    x_boot, y_boot = resample(x, y)
    model.fit(x_boot, y_boot)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
  return accuracy_scores

svm_accuracy_scores = bootstrap_model(svm, x_train, y_train)
lr_accuracy_scores = bootstrap_model(LR, x_train, y_train)
knn_accuracy_scores = bootstrap_model(knn, x_train, y_train)
```

```python
print("SVM Accuracy :", np.mean(svm_accuracy_scores))
```

```python
print("KNN Accuracy :", np.mean(knn_accuracy_scores))
```

```python
print("Logistic Regression Accuracy:", np.mean(lr_accuracy_scores))
```

#**SVM**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(svm_accuracy_scores, alpha=0.5, label='SVM')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.title('Bootstrap Distribution of Accuracy Scores')
plt.legend()
plt.show()
```
