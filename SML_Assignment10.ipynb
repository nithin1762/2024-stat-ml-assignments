```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/breast_cancer_survival.csv')
data.head()
```

```python
data=data.dropna()
data = data.replace({'FEMALE':0, 'MALE':1, 'Negative':0, 'Positive':1, 'Alive':1, 'Dead':0, 'I':1, 'II':2,'III':3, 'Infiltrating Ductal Carcinoma':1, 'Infiltrating Lobular Carcinoma':2, 'Mucinous Carcinoma':3, 'Lumpectomy':1, 'Modified Radical Mastectomy':2, 'Simple Mastectomy':3, 'Other':4})
data
```

```python
x = data.drop(['Patient_Status', 'Date_of_Surgery', 'Date_of_Last_Visit'], axis=1) #features
y = data['Patient_Status'] #target
```

```python
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(x, y, test_size=0.2, random_state=42)
x_train2, x_test2, y_train2, y_test2 = train_test_split(x, y, test_size=0.25, random_state=42)
x_train3, x_test3, y_train3, y_test3 = train_test_split(x, y, test_size=0.3, random_state=42)
x_train4, x_test4, y_train4, y_test4 = train_test_split(x, y, test_size=0.35, random_state=42)
x_train = [x_train1, x_train2, x_train3, x_train4]
y_train = [y_train1, y_train2, y_train3, y_train4]
x_test = [x_test1, x_test2, x_test3, x_test4]
y_test = [y_test1, y_test2, y_test3, y_test4]
test_size = np.array([0.2, 0.25, 0.3, 0.35])
random_state = [42, 42, 42, 42]
```

# **KNN model**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
k = range(1,51)
k = k[::2]
for j in k:
  knn = KNeighborsClassifier(n_neighbors=j)
  print(f"******** Number of neighbors: {j} ********")
  for i in range(4):
    print(f"Test size {i+1}: {test_size[i]}")
    knn.fit(x_train[i], y_train[i])
    y_pred = knn.predict(x_test[i])
    accuracy = accuracy_score(y_test[i], y_pred)
    print(f"Accuracy: {accuracy}")
    print("-------------------------------------------------------------")
  print("==============================================================================")
  print("\n\n")
```

# **SVM model**

```python
from sklearn.svm import SVC
import numpy as np

accuracy_list = []
SVM = SVC(kernel='linear')
for i in range(4):
  print(f"Test size {i+1}: {test_size[i]}")
  SVM.fit(x_train[i], y_train[i])
  y_pred = SVM.predict(x_test[i])
  accuracy = accuracy_score(y_test[i], y_pred)
  accuracy_list.append(accuracy)
  print(f"Accuracy: {accuracy}")
  print("-------------------------------------------------------------")

accuracy_list = np.array(accuracy_list)
```

```python
import matplotlib.pyplot as plt
plt.plot(test_size, accuracy_list)
plt.scatter(test_size, accuracy_list, color='red')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Test Size')
plt.show()
```
